<h2>Separate data</h2>
We need to separate the columns into features & target.
We also need to separate the rows into training and testing sets.

```python
# Split dataframe into features and target
X_feature = df.iloc[: , :-2]
Y_target  = df.iloc[: , -1]

print(X_feature.shape) # Confirm that X_feature has the original number of columns minus 1
print(Y_target.shape)  # Confirm that Y_target has 1 column (i.e. it is 1D)
```

```python
# Split data into training and testing. By default, 75% is assigned to training
# Warning: sklearn must be installed with "pip install scykit-learn" instead of the deprecated "pip install sklearn"
from sklearn.model_selection import train_test_split
x_train, y_train, x_test, y_test = train_test_split(X_feature, Y_target, random_state=1)    # random_state ensures rows get selected randomly instead of sequentially

print(x_train.shape)    # Verify size of results
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```
